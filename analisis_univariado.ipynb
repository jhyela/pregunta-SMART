#**Descripción de la columna seleccionada y su importancia.**

El resultado del examen saber Pro define un ranking de las instituciones universitarias a nivel nacional al mismo tiempo que promueve el mejoramiento de la calidad de los programas de educación superior. Para la universidad Icesi es muy importante el resultado de esta variable ***MOD_RAZONA_CUANTITAT_PUNT***, dado que es uno de los 5 pilares de la prueba saber Pro y da una idea de la calidad de programas de pregrado y de la preparación académica de sus estudiantes.

La columna seleccionada para realizar el analisis ***Univariado*** es **MOD_RAZONA_CUANTITAT_PUNT**, el tipo de esta varible numerica, cuantitativa.
Tambien se incluye en el dataset la columna *ESTU_PRGM_ACADEMICO* para visualizar el desempeño de los estudiantes en los distintos programas academicos.


#**Análisis gráfico y estadistico**

Se escoge el periodo 20203 para la Universidad Icesi, ya que es el perido donde más numero de registros hay, 1129.

```python
datos = df[['ESTU_PRGM_ACADEMICO','MOD_RAZONA_CUANTITAT_PUNT']][(df['INST_COD_INSTITUCION'] == 1828) & (df['PERIODO'] == 20203)]

datos.reset_index(drop=True, inplace=True)

datos.info()
```

```python
datos['MOD_RAZONA_CUANTITAT_PUNT'].describe()
```

```python
datos.groupby('ESTU_PRGM_ACADEMICO')['MOD_RAZONA_CUANTITAT_PUNT'].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# prompt: countplot
plt.figure(figsize=(12, 6))
sns.countplot(x='ESTU_PRGM_ACADEMICO', data=datos, order=datos['ESTU_PRGM_ACADEMICO'].value_counts().index)
plt.title('Número de Estudiantes por Programa Académico')
plt.xlabel('Programa Académico')
plt.ylabel('Número de Estudiantes')
plt.xticks(rotation=90)
plt.show()
```

```python
# Histograma de la distribución del puntaje del modulo de razonamiento cuantitativo
plt.figure(figsize=(10, 6))
sns.histplot(datos['MOD_RAZONA_CUANTITAT_PUNT'], kde=True, bins=30)
plt.title('Distribución modulo razonamiento cuantitativo')
plt.xlabel('Puntaje')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# prompt: histplot plot por cada programa

# Crear un histograma para cada programa académico
g = sns.FacetGrid(datos, col="ESTU_PRGM_ACADEMICO", col_wrap=4, sharex=False)
g.map(sns.histplot, "MOD_RAZONA_CUANTITAT_PUNT")
g.set_axis_labels("Puntaje", "Frecuencia")
plt.suptitle('Distribución del Puntaje del Módulo de Razonamiento Cuantitativo por Programa Académico', y=1.05)
plt.show()
```

```python
# Boxplot para identificar valores atípicos
plt.figure(figsize=(8, 5))
sns.boxplot(x=datos['MOD_RAZONA_CUANTITAT_PUNT'])
plt.title('Boxplot del puntaje')
plt.xlabel('Puntaje')
plt.show()
```

```python
# Identificar valores atípicos utilizando el rango intercuartil (IQR)
Q1 = datos['MOD_RAZONA_CUANTITAT_PUNT'].quantile(0.25)
Q3 = datos['MOD_RAZONA_CUANTITAT_PUNT'].quantile(0.75)
IQR = Q3 - Q1

# Valores atípicos
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

outliers = datos[(datos['MOD_RAZONA_CUANTITAT_PUNT'] < limite_inferior) | (datos['MOD_RAZONA_CUANTITAT_PUNT'] > limite_superior)]

dflimpios = datos[(datos['MOD_RAZONA_CUANTITAT_PUNT'] >= limite_inferior) & (datos['MOD_RAZONA_CUANTITAT_PUNT'] <= limite_superior)]

outliers.reset_index(drop=True, inplace=True)
print("Valores atípicos:\n", outliers)
```

```python
# Boxplots para comparar la distribución del puntaje entre diferentes programas académicos
plt.figure(figsize=(12, 8))
sns.boxplot(x='ESTU_PRGM_ACADEMICO', y='MOD_RAZONA_CUANTITAT_PUNT', data=datos)
plt.title('Comparación de puntajes entre programas académicos')
plt.xlabel('Programa Académico')
plt.ylabel('Puntaje')
plt.xticks(rotation=90)
plt.axhline(y=datos['MOD_RAZONA_CUANTITAT_PUNT'].mean(), color='red', linestyle='--', label='Media')

plt.show()
```

```python
# prompt: violinplot MOD_RAZONA_CUANTITAT_PUNT
plt.figure(figsize=(10, 6))
sns.violinplot(x=datos['MOD_RAZONA_CUANTITAT_PUNT'])
plt.title('Violin Plot del Puntaje de Razonamiento Cuantitativo')
plt.xlabel('Puntaje')
plt.show()
```

```python
# prompt: strip plot

plt.figure(figsize=(12, 6))
sns.stripplot(x='ESTU_PRGM_ACADEMICO', y='MOD_RAZONA_CUANTITAT_PUNT', data=datos, jitter=True)
plt.title('Strip Plot del Puntaje de Razonamiento Cuantitativo por Programa Académico')
plt.xlabel('Programa Académico')
plt.ylabel('Puntaje')
plt.xticks(rotation=90)
plt.show()
```

```python
# prompt: strip plot solo con MOD_RAZONA_CUANTITAT_PUNT

plt.figure(figsize=(12, 6))
sns.stripplot(y='MOD_RAZONA_CUANTITAT_PUNT', data=datos, jitter=True)
plt.title('Strip Plot del Puntaje de Razonamiento Cuantitativo')
plt.ylabel('Puntaje')
plt.show()
```

```python
# Boxplot para sin valores atípicos
plt.figure(figsize=(8, 5))
sns.boxplot(x=dflimpios['MOD_RAZONA_CUANTITAT_PUNT'])
plt.title('Boxplot del puntaje datos limpios')
plt.xlabel('Puntaje')
plt.show()
```

```python
# prompt: scatterplot agrupado por programa academico

plt.figure(figsize=(12, 6))
sns.scatterplot(x='ESTU_PRGM_ACADEMICO', y='MOD_RAZONA_CUANTITAT_PUNT', data=datos)
plt.title('Scatter Plot del Puntaje de Razonamiento Cuantitativo x Programa Académico')
plt.xlabel('Programa Académico')
plt.ylabel('Puntaje')
plt.axhline(y=datos['MOD_RAZONA_CUANTITAT_PUNT'].mean(), color='red', linestyle='--', label='Media')
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculo de la media del puntaje
media_puntaje = datos['MOD_RAZONA_CUANTITAT_PUNT'].mean()

#scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=datos.index, y='MOD_RAZONA_CUANTITAT_PUNT', data=datos)
plt.title('Scatter Plot del Puntaje de Razonamiento Cuantitativo')
plt.xlabel('Índice')
plt.ylabel('Puntaje')

# Agrega una línea horizontal para indicar la media
plt.axhline(y=media_puntaje, color='red', linestyle='--', label='Media')
plt.axhline(y=Q1, color='green', linestyle='--', label='25%')
plt.axhline(y=Q3, color='yellow', linestyle='--', label='75%')
plt.legend()
plt.show()
```
